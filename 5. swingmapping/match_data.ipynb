```python
import os
import pandas as pd
import json

def get_MatchedFiles():
        
    # File directories 
    input_DIR = r'C:\Users\James\Desktop\Raw\logRaws'
    result_DIR = r'C:\Users\James\Desktop\Raw\userInfo'

    # List all files in the directory
    input_files = os.listdir(input_DIR)
    result_files = os.listdir(result_DIR)

    # Match input and output files based on the common timestamp prefix 
    matched_files = {}

    for input_file in input_files:
        # Extract the common timestamp prefix and construct .json file
        common_prefix = '_'.join(input_file.split('_')[:-1]) + '_'
        output_file = common_prefix + 'userinfo.json'

        # Check if the matching output file exists
        if output_file in result_files:
            matched_files[input_file] = output_file

    print("Number of Matched Files : %s" %len(matched_files))
    return matched_files

files = get_MatchedFiles()
```

```python
import os

def delete_unmatched_files(result_DIR, matched_files):
    # List all files in the result directory
    result_files = os.listdir(result_DIR)

    deleted_count = 0
    # Delete files that do not have matching files
    for result_file in result_files:
        if result_file not in matched_files.values():
            file_path = os.path.join(result_DIR, result_file)
            os.remove(file_path)
            print(f"Deleted {file_path}")
            deleted_count = deleted_count + 1

    print("Number of Deleted userInfo Files : %s" % deleted_count)

matched_files = get_MatchedFiles()
result_DIR = r'C:\Users\James\Desktop\Raw\userInfo'

delete_unmatched_files(result_DIR, matched_files)
```
