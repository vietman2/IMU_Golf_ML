```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# prepare data

path = '../data/Raw/example.csv'

raw = pd.read_csv(path, header=None)

noisy = raw.iloc[:, [16]]
fil = raw.iloc[:, [19]]

def plot(noisy, filtered, ma):
    plt.figure(figsize=(20, 10))
    plt.plot(noisy, label='non-filtered')
    plt.plot(filtered, label='filtered')
    plt.plot(ma, label='moving averages')
    plt.legend()
    plt.show()

def moving_averages(data, window_size):
    return data.rolling(window=window_size).mean() 

plot(noisy, fil, moving_averages(noisy, 3))
```
